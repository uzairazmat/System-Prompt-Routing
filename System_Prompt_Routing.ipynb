```python
from sentence_transformers import SentenceTransformer

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Create embeddings for each prompt
embeddings = {
    key: model.encode(text, convert_to_tensor=True)
    for key, text in multimodal_prompts.items()
}

```

```python
import numpy as np

# Convert tensor embeddings to numpy float32 and keep the same order as keys
keys = list(multimodal_prompts.keys())           # order matters
vectors = []
for k in keys:
    t = embeddings[k]            # this is a torch tensor (convert_to_tensor=True)
    vec = t.cpu().numpy()        # to numpy
    vec = vec.astype('float32')  # FAISS requires float32
    vectors.append(vec)
```

```python
X = np.vstack(vectors)  # shape (n_prompts, dim)
# normalize for cosine similarity (so inner product == cosine)
norms = np.linalg.norm(X, axis=1, keepdims=True)
# avoid division by zero
norms[norms == 0] = 1e-10
X = X / norms
print("Vectors shape:", X.shape)
```

#2) Create a FAISS index (exact, good for small N)

Why: IndexFlatIP does exact inner-product search — use with normalized vectors for cosine.

```python
import faiss

dim = X.shape[1]
index = faiss.IndexFlatIP(dim)   # inner-product index
index.add(X)                    # add all vectors at once
print("Index size:", index.ntotal)
```

```python
import pickle

# Save FAISS index binary
faiss.write_index(index, "prompt_index.faiss")

# Save metadata (keys and prompts)
meta = {"keys": keys, "prompts": multimodal_prompts}
with open("prompt_meta.pkl", "wb") as f:
    pickle.dump(meta, f)
```

```python
# At startup
index = faiss.read_index("prompt_index.faiss")
with open("prompt_meta.pkl", "rb") as f:
    meta = pickle.load(f)
keys = meta["keys"]
prompts = meta["prompts"]
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

def query_router(query, k=3, score_threshold=0.4):
    q_emb = model.encode([query], convert_to_numpy=True).astype('float32')
    q_emb = q_emb / np.linalg.norm(q_emb, axis=1, keepdims=True)
    D, I = index.search(q_emb, k)   # D = similarity scores (inner product), I = indices
    # D shape (1,k), I shape (1,k)
    results = []
    for score, idx in zip(D[0], I[0]):
        if idx == -1:
            continue
        key = keys[idx]
        prompt_text = prompts[key]
        results.append({"key": key, "prompt": prompt_text, "score": float(score)})
    # Apply a threshold for best match fallback
    best = results[0] if results and results[0]["score"] >= score_threshold else None
    return {"query": query, "candidates": results, "best": best}

# Example usage
out = query_router("How do I use inheritance in Python?", k=3, score_threshold=0.35)
print(out)
```
